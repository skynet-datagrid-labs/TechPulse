# TechPulse Exploratory Data Analysis

This notebook explores class distributions, feature relationships, and missing values for the TechPulse ML pipeline.

```python
from pathlib import Path

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

BASE_DIR = Path(".").resolve().parents[0]
DATA_DIR = BASE_DIR / "data"
REPORTS_DIR = BASE_DIR / "reports"
FIGURES_DIR = REPORTS_DIR / "figures"
FIGURES_DIR.mkdir(parents=True, exist_ok=True)

feature_path = DATA_DIR / "feature_matrix.csv"
if not feature_path.exists():
    raise FileNotFoundError("Run src/features.py to generate data/feature_matrix.csv")

features = pd.read_csv(feature_path)
features.head()
```

```python
# Class distribution bar chart
plt.figure(figsize=(6, 4))
class_counts = features["label"].value_counts().reset_index()
sns.barplot(data=class_counts, x="label", y="count", hue="label", legend=False)
plt.title("Class Distribution")
plt.xlabel("Class")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig(FIGURES_DIR / "class_distribution.png", dpi=150)
plt.close()
```

```python
# Correlation heatmap
numeric_features = features.select_dtypes(include="number")
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_features.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.tight_layout()
plt.savefig(FIGURES_DIR / "correlation_heatmap.png", dpi=150)
plt.close()
```

```python
# Box plots of each feature by class
for column in numeric_features.columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=features, x="label", y=column)
    plt.title(f"{column} by Class")
    plt.tight_layout()
    plt.savefig(FIGURES_DIR / f"boxplot_{column}.png", dpi=150)
    plt.close()
```

```python
# Top 10 technologies per class
top_tech = (
    features.groupby("label")
    .head(10)
    .loc[:, ["technology", "label"]]
)
top_tech
```

```python
# Missing value analysis
missing_counts = features.isna().sum().sort_values(ascending=False)
plt.figure(figsize=(8, 4))
sns.barplot(x=missing_counts.index, y=missing_counts.values)
plt.xticks(rotation=45, ha="right")
plt.title("Missing Values by Feature")
plt.ylabel("Missing Count")
plt.tight_layout()
plt.savefig(FIGURES_DIR / "missing_values.png", dpi=150)
plt.close()
```

```python
# Feature distribution histograms
for column in numeric_features.columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(features[column], kde=True)
    plt.title(f"Distribution of {column}")
    plt.tight_layout()
    plt.savefig(FIGURES_DIR / f"hist_{column}.png", dpi=150)
    plt.close()
```
